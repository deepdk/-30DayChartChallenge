```python
fig = plt.figure(
    FigureClass=Waffle,
    rows=15,
    values=[98,96,94,58,54,54,40,40,30,18],
    cmap_name="tab10",
    icons=['fish','fish','fish','fish','fish','fish','fish','fish','fish','fish'],
    font_size=20,
    icon_legend=True,
    figsize = (20,20),
    legend={
        'labels': ['Crappie', 'Channel catfish', 'Herring','Rock Bass and Crappie','Smallmouth Bass','American Eel','Bullhead','Drum','White bass','Sheepshead'], 
        'loc': 'upper left', 
        'bbox_to_anchor': (1, 1),
        'fontsize' : 'xx-large'
    }
)

plt.title("Least Found Fish Species in US and Canada", fontsize = 30)
txt="Data Source : Commercial Fishing, #TidyTuesday/Week 24/2021"
plt.figtext(0.4, 0.33, txt, wrap=True, horizontalalignment='center', fontsize=15)
```
